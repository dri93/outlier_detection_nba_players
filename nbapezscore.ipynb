```python
import pandas as pd
import seaborn as sns
from scipy.spatial.distance import pdist, squareform
```

```python
df = pd.read_csv("/Users/adriannabowen/CS 663 - Data Mining/Final Project/cleaned datasets/player_regular_season.csv")
df2 = pd.read_csv("/Users/adriannabowen/CS 663 - Data Mining/Final Project/cleaned datasets/team_season.csv")
df.head()
df2.head()
```

```python
playerstats = df[(df.year == 2004)]
teamstats = df2[(df2.year == 2004)]
```

```python
combine = pd.merge(playerstats, teamstats, how="left", on="team")
combine.head()
```

```python
dff = combine.groupby(["team"]).minutes.sum().reset_index()
dff.head()
```

```python
totalstats = pd.merge(combine, dff, how="left", on="team")
totalstats.rename(columns={'minutes_y':'tm_mp'}, inplace=True)
totalstats.head()
```

```python
drop = ['Unnamed: 0_x','leag_x','Unnamed: 0_y','year_y', 'leag_y','pace','won','lost','oreb','dreb','pf','tpa','tpm','o_3pm','o_3pa','d_oreb','d_dreb']
stat = totalstats.drop(columns=drop)
stat.head()
```

```python
stat['usage'] = 100 * ((stat.fga+.44*stat.fta+stat.turnover)* (stat.tm_mp/5))/(stat.minutes_x*((stat.o_fga+stat.d_fga)+.44*(stat.o_fta+stat.d_fta)+(stat.o_to+stat.d_to)))
stat.head()
```

```python
stat['cumulative_efficiency'] = (stat.pts + stat.reb + stat.asts + stat.stl + stat.blk - (stat.fga-stat.fgm) - (stat.fta-stat.ftm) - stat.turnover)
stat['players_efficiency'] = (stat.pts + stat.reb + stat.asts + stat.stl + stat.blk - (stat.fga-stat.fgm) - (stat.fta-stat.ftm) - stat.turnover)/stat.gp
```

```python
mean = stat["players_efficiency"].mean()
std = stat["players_efficiency"].std()
print('mean is', mean)
print('standard deviation is', std)
```

```python
threshold = 2
outlier = []
for i in stat["players_efficiency"]:
    z = (i-mean)/std
    if z > threshold:
        outlier.append(i)
print('outlier in dataset is', outlier)
```
