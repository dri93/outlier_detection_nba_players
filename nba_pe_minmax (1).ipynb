```python
playerstats = df#[(df.year == 2004)]
teamstats = df2#[(df2.year == 2004)]
playerstats.shape, teamstats.shape
playerstats
```

Combine the playerstats and teamstats dataframes. 

```python
combine = pd.merge(playerstats, teamstats, how="inner", on=["team", "year", "leag"])
combine.sort_values('ilkid')
```

```python
combine[combine.isnull().any(axis=1)]
```

```python
dff = combine.groupby(["team"]).minutes.sum().reset_index()
dff.head()
```

```python
totalstats = pd.merge(combine, dff, how="left", on="team")
totalstats.rename(columns={'minutes_y':'tm_mp'}, inplace=True)
totalstats.head()
```

```python
drop = ['pace','won','lost','oreb','dreb','pf','tpa','tpm','o_3pm','o_3pa','d_oreb','d_dreb']
stat = totalstats.drop(columns=drop)
stat.head()
```

# We created a version of our stat dataframe with normalized columns.

```python
stat_norm = stat.copy()
columns_to_normalize = ['fga', 'fta', 'turnover',
                        'tm_mp', 'minutes_x', 'o_fga',
                        'd_fga', 'o_fta', 'd_fta',
                        'o_to', 'd_to', 'pts', 'reb',
                        'asts', 'stl', 'blk', 'fgm',
                        'ftm']
for column in columns_to_normalize:
    stat_norm.loc[:, column] -= stat_norm[column].min()
    stat_norm.loc[:, column] /= stat_norm[column].max()
```

# We calculated Usage as well as Player Efficiency Rating (PER) for each player, although we did not "use" Usage in our analysis later on.

```python
def usage(stat):
    return 100 * (((stat.fga+.44*stat.fta+stat.turnover)*(stat.tm_mp/5))
                  /(stat.minutes_x*((stat.o_fga+stat.d_fga)+
                                    .44*(stat.o_fta+stat.d_fta)+
                                    (stat.o_to+stat.d_to))))

def efficiency(stat, is_cumulative):
    eff = (stat.pts + stat.reb + stat.asts + stat.stl + stat.blk -
           (stat.fga-stat.fgm) - (stat.fta-stat.ftm) - stat.turnover)
    if is_cumulative:
        return eff
    else:
        return eff/stat.gp
```

```python
stat['usage'] = usage(stat)
stat.sort_values('usage')
```

```python
stat_norm['usage'] = usage(stat_norm)
stat_norm.sort_values('usage')
```

```python
stat['cumulative_efficiency'] = efficiency(stat, True)
stat['players_efficiency'] = efficiency(stat, False)
stat
```

```python
stat_norm['cumulative_efficiency'] = efficiency(stat_norm, True)
stat_norm['players_efficiency'] = efficiency(stat_norm, False)
stat_norm
```

# We looked at outliers using both the standard PER metric and the one with normalized components, and check to see the extent of overlap. There was about 10% overlap.

```python
def get_outliers(stat, variable, threshold=1):
    mean = stat[variable].mean()
#     std = stat[variable].std()
    mean_abs_diff = (stat[variable] - mean).abs().mean() # more robust
    outlier = []
    for x in stat[variable]:
        z = (x - mean) / mean_abs_diff
        if abs(z) > threshold:
            outlier.append(x)
    return stat.loc[stat[variable].abs().isin(outlier)]
```

```python
unnorm_outliers = get_outliers(stat, "players_efficiency")
norm_outliers = get_outliers(stat_norm, "players_efficiency")

df = pd.merge(unnorm_outliers, norm_outliers, how='inner', on=['ilkid'])
```

```python
both_lists = (df['firstname_x_x']+' '+df['lastname_x_x']+', aka '+df['firstname_y_y']+' '+df['lastname_y_y']).unique()
both_lists.sort()
for p in both_lists:
    print(p)
```
